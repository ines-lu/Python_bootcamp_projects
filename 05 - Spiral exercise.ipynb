# Spiral exercise

Read in the file "spiral_data.txt" (attached). Each line of this file contains the x and y coordinates of a point in a 2D plane. I want you to read this file, save the x and y values in an appropriate data structure and then plot x values versus the y values (scatter plot).

```python
# read the file
myfile = open(file='spiral_data.txt', mode="r")
file_output = myfile.readlines()
myfile.close()

xy_list = []
for i in file_output:
    xy_list.append( i.strip("\n"))
print(type(xy_list))
print(xy_list)

x = []
y = []
for i in xy_list:
    i1 = i.split(",")
    x.append( float(i1[0]) ) 
    y.append( float(i1[-1])) 
```

```python
# save x and y
xy_list = []
for i in file_output:
    xy_list.append( i.strip("\n"))

x = []
y = []
for i in xy_list:
    i1 = i.split(",")
    x.append( float(i1[0]) ) 
    y.append( float(i1[-1])) 
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.scatter(x,y)

ax.set_title("my plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
    
```
